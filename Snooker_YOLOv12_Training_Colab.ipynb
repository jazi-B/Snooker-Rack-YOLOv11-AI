### Step 3: Run YOLOv12 Production Training (Auto GPU / CPU Selection)

```python
import torch
from ultralytics import YOLO

# Auto-detect GPU or CPU
target_device = 0 if torch.cuda.is_available() else 'cpu'
print(f'[*] Active Training Device: {target_device} (CUDA Available: {torch.cuda.is_available()})')

# Load Pretrained YOLOv12 Nano Weights
model = YOLO('yolo12n.pt')

# Launch Multi-Angle Production Training
results = model.train(
    data='data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='snooker_rack_yolov12_colab',
    project='runs/detect',
    exist_ok=True,
    device=target_device,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=25.0,
    translate=0.15,
    scale=0.35,
    perspective=0.001,
    mosaic=0.8,
    mixup=0.15,
    patience=20
)
print('[+] Training Complete!')
```

### Step 4: Evaluate Validation Metrics (Precision, Recall, mAP50)

```python
best_model = YOLO('runs/detect/snooker_rack_yolov12_colab/weights/best.pt')
metrics = best_model.val(data='data.yaml', split='val')

print('=' * 50)
print(f'Precision: {metrics.box.mp*100:.2f}%')
print(f'Recall:    {metrics.box.mr*100:.2f}%')
print(f'mAP@50:    {metrics.box.map50*100:.2f}%')
print(f'mAP@50-95: {metrics.box.map*100:.2f}%')
```
